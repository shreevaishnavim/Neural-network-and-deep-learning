```
# Install libraries
!pip install face-recognition opencv-python

# Imports
import cv2
import face_recognition
import numpy as np
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
from google.colab import files

# ==============================
# Function to capture photo
# ==============================
def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
        async function takePhoto(quality) {
            const div = document.createElement('div');
            const button = document.createElement('button');
            button.textContent = '📸 Capture';
            div.appendChild(button);

            const video = document.createElement('video');
            video.style.display = 'block';

            const stream = await navigator.mediaDevices.getUserMedia({video: true});
            document.body.appendChild(div);
            div.appendChild(video);
            video.srcObject = stream;
            await video.play();

            await new Promise((resolve) => button.onclick = resolve);

            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            canvas.getContext('2d').drawImage(video, 0, 0);

            stream.getTracks().forEach(track => track.stop());
            div.remove();

            return canvas.toDataURL('image/jpeg', quality);
        }
    ''')
    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

# ==============================
# Upload your face image
# ==============================
print("Upload your image (my_face.jpg)")
uploaded = files.upload()

# Load your face
# Assuming the uploaded file is named 'my_face.jpg.jpeg' or similar
# You might need to adjust the filename here if it's different
uploaded_filename = list(uploaded.keys())[0]
my_image = face_recognition.load_image_file(uploaded_filename)
encodings = face_recognition.face_encodings(my_image)

if len(encodings) == 0:
    print("❌ No face found in uploaded image")
else:
    my_encoding = encodings[0]
    print("✅ Face loaded successfully")

    # ==============================
    # Capture image from webcam
    # ==============================
    print("\nClick '📸 Capture' when ready to take a photo from your webcam.")
    photo = take_photo()
    print("📸 Photo captured")

    # ==============================
    # Recognize face
    # ==============================
    test_image = face_recognition.load_image_file(photo)
    test_encodings = face_recognition.face_encodings(test_image)

    if len(test_encodings) == 0:
        print("❌ No face detected in webcam image")
    else:
        test_encoding = test_encodings[0]

        result = face_recognition.compare_faces([my_encoding], test_encoding)

        if True in result:
            print("✅ Present (You)")
        else:
            print("❌ Not You")
```
