```python
import math

x = [1, 0, 1]
h1 = [0.5, -0.2, 0.3]
h2 = [0.9, 0.1, -0.5]
out = [0.7, 0.2]

relu = lambda z: max(0, z)
sigmoid = lambda z: 1 / (1 + math.exp(-z))

h1_out = relu(x[0]*h1[0] + x[1]*h1[1] + x[2]*h1[2])
h2_out = relu(x[0]*h2[0] + x[1]*h2[1] + x[2]*h2[2])

out_sum = h1_out*out[0] + h2_out*out[1]
y = sigmoid(out_sum)

print("Output =", y)
print("Spam" if y > 0.5 else "Not Spam")
```

```python
import math
x1, x2, x3 = 1, 0, 1  
W1 = [0.5, -0.2, 0.3]  
b1 = 0.0
W2 = [0.4, 0.1, -0.5]  # h2 weights
b2 = 0.0
W_out_h1 = 0.7  
W_out_h2 = 0.2  
b_out = 0.2     
def relu(x):
    return max(0, x)
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
sum_h1 = x1*W1[0] + x2*W1[1] + x3*W1[2] + b1
h1 = relu(sum_h1)  # 0.8

sum_h2 = x1*W2[0] + x2*W2[1] + x3*W2[2] + b2
h2 = relu(sum_h2)  # 0.0
sum_out = h1*W_out_h1 + h2*W_out_h2 + b_out
output = sigmoid(sum_out)

print("Hidden neuron h1 (ReLU):", h1)
print("Hidden neuron h2 (ReLU):", h2)
print("Final Output (Sigmoid):", round(output, 2))

if output > 0.5:
    print("Prediction: SPAM MAIL")
else:
    print("Prediction: NOT SPAM")
```
