1. What is the syntax for creating a tensor.

```
The syntax varies by framework:

PyTorch: torch.tensor(data) or factory functions like torch.zeros(), torch.ones(), torch.randn()

TensorFlow: tf.constant(data) or tf.Variable(data)

NumPy: np.array(data) (technically arrays, not tensors with gradients)

Key Components:

Data (list, array, or scalar)

Optional: dtype, device (CPU/GPU), requires_grad (PyTorch)
```

2. what are the different tensor operations



```
Tensor operations can be categorized as:

1.Creation Operations:

zeros(), ones(), randn(), arange(), linspace()

2.Mathematical Operations:

Element-wise: +, -, *, /, exp(), log()

3.Matrix operations: matmul(), dot(), transpose()

Reduction: sum(), mean(), max(), min()

4.Shape Operations:

reshape(), view(), permute(), squeeze(), unsqueeze()

5.Indexing & Slicing:

tensor[i, j], tensor[:, 2:5], masked_select()

6.Statistical Operations:

std(), var(), norm(), argmax()

```

8.Compute matrix multiplication between tensors of shape (3,4) and (4,2).



```
import torch

# Create tensors
tensor_A = torch.randn(3, 4)  # Shape (3, 4)
tensor_B = torch.randn(4, 2)  # Shape (4, 2)

print("Tensor A (3x4):\n", tensor_A)
print("\nTensor B (4x2):\n", tensor_B)

# Matrix multiplication
# Method 1: torch.matmul()
result_matmul = torch.matmul(tensor_A, tensor_B)
print("\nResult using matmul() (3x2):\n", result_matmul)

# Method 2: @ operator
result_operator = tensor_A @ tensor_B
print("\nResult using @ operator (3x2):\n", result_operator)

# Method 3: torch.mm() (only for 2D tensors)
result_mm = torch.mm(tensor_A, tensor_B)
print("\nResult using mm() (3x2):\n", result_mm)

print("\nAll methods equal:", torch.allclose(result_matmul, result_operator) and
      torch.allclose(result_matmul, result_mm))
```

9.Find the transpose of a 3D tensor.



```
import torch

# Create a 3D tensor (example: 2 batches, 3 rows, 4 columns)
tensor_3d = torch.tensor([
    [[1, 2, 3, 4],   # Batch 0, Row 0
     [5, 6, 7, 8],   # Batch 0, Row 1
     [9, 10, 11, 12]], # Batch 0, Row 2

    [[13, 14, 15, 16],  # Batch 1, Row 0
     [17, 18, 19, 20],  # Batch 1, Row 1
     [21, 22, 23, 24]]  # Batch 1, Row 2
])

print("Original 3D Tensor shape:", tensor_3d.shape)  # (2, 3, 4)
print("Original tensor:\n", tensor_3d)

# Method 1: Transpose specific dimensions (swap dimensions 1 and 2)
transposed = tensor_3d.transpose(1, 2)  # Swap rows and columns
print("\nTransposed (swap dims 1&2) shape:", transposed.shape)  # (2, 4, 3)
print("Transposed tensor:\n", transposed)

# Method 2: Using permute for any dimension order
permuted = tensor_3d.permute(0, 2, 1)  # Equivalent to transpose(1, 2)
print("\nPermuted shape:", permuted.shape)

# Method 3: Transpose each 2D matrix individually
print("\nTransposing each 2D slice individually:")
for i in range(tensor_3d.shape[0]):
    print(f"\nBatch {i} original:\n", tensor_3d[i])
    print(f"Batch {i} transposed:\n", tensor_3d[i].T)  # .T works for 2D tensors
```

10.Implement softmax using tensor operations (no built-in function).



```
import numpy as np

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

#Simple array
print("Simple array")
x1 = np.array([1.0, 2.0, 3.0])
result1 = softmax(x1)
print(f"Input: {x1}")
print(f"Output: {result1}")
print(f"Sum: {np.sum(result1):.6f}")  # Should sum to 1
print()
```

11.Normalize a tensor so that values lie between 0 and 1.



```
import numpy as np

# 2-mark solution
def normalize(tensor):
    return (tensor - tensor.min()) / (tensor.max() - tensor.min())

# Test and print
x = np.array([10, 20, 30, 40, 50])
result = normalize(x)

print("Input tensor:", x)
print("Normalized:", result)
print("Min value:", result.min())
print("Max value:", result.max())
```

12.Write code to compute gradients of a tensor operation (PyTorch).

```
import torch

# Create a tensor with requires_grad=True for gradient computation
x = torch.tensor([2.0], requires_grad=True)

# Define an operation: y = x^2 + 3x + 1
y = x**2 + 3*x + 1

# Compute gradients
y.backward()

# Print results
print("x =", x.item())
print("y = x² + 3x + 1 =", y.item())
print("Gradient dy/dx =", x.grad.item())
print("Analytical derivative: 2x + 3 =", 2*x.item() + 3)
```

13.Convert an image tensor (H, W, C) to (C, H, W).

```
import torch

# Create a random image tensor: (Height, Width, Channels)
# Simulating a 4x3 RGB image
image_hwc = torch.randn(4, 3, 3)  # H=4, W=3, C=3
print("Original shape (H, W, C):", image_hwc.shape)

# Convert to (C, H, W) using permute
image_chw = image_hwc.permute(2, 0, 1)
print("Converted shape (C, H, W):", image_chw.shape)
print()

# Alternative method using .permute
print("Using permute(2, 0, 1):")
print("Input shape:", image_hwc.shape)
print("Output shape:", image_hwc.permute(2, 0, 1).shape)
```

15.Difference between view, reshape, and transpose in tensors.



```
view(): Changes shape without copying data (memory efficient). Requires tensor to be contiguous.

reshape(): Changes shape may copy data if needed. Works on both contiguous/non-contiguous tensors.

transpose(): Swaps dimensions (e.g., rows ↔ columns). Changes stride pattern, not memory layout.

```

16.What is contiguous memory in tensor operations?



```
Contiguous memory means tensor elements are stored sequentially in memory without gaps. Operations on contiguous tensors are faster because they enable:

1.Better cache utilization

2.Vectorized operations (SIMD)

3.Direct memory access patterns
Example: [1,2,3,4,5] is contiguous; [[1,2],[3,4]] after transpose may not be.

```

17.Why is broadcasting efficient?



```
Broadcasting avoids actual data duplication. Instead of copying data to match shapes:

1.Uses stride tricks to virtually repeat data

2.Reduces memory usage

3.Enables vectorized operations
Example: Adding [1,2,3] to each row of matrix uses same 3 values without copying.

```

20.Implement cosine similarity using tensor operations.



```
import torch

def cosine_similarity(a, b):
    # Normalize vectors
    a_norm = a / torch.norm(a)
    b_norm = b / torch.norm(b)
    # Dot product
    return torch.dot(a_norm, b_norm)

# Test
v1 = torch.tensor([1.0, 2.0, 3.0])
v2 = torch.tensor([4.0, 5.0, 6.0])
print("Cosine similarity:", cosine_similarity(v1, v2).item())
```
