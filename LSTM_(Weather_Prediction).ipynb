```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from sklearn.preprocessing import MinMaxScaler

# Load Dataset
data = pd.read_csv('/content/DailyDelhiClimateTest.csv')

# Use 'meantemp' column
temp = data['meantemp'].values.reshape(-1,1)

# Normalize Data
scaler = MinMaxScaler()
temp_scaled = scaler.fit_transform(temp)

# Create Sequences (5 days)
X, y = [], []
for i in range(5, len(temp_scaled)):
    X.append(temp_scaled[i-5:i])
    y.append(temp_scaled[i])

X, y = np.array(X), np.array(y)

# Split Data (80% training)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Build LSTM Model
model = Sequential()
model.add(Input(shape=(X.shape[1], 1))) # Added Input layer to address warning
model.add(LSTM(50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Train Model
model.fit(X_train, y_train, epochs=10, batch_size=16)

# Predict
predicted = model.predict(X_test)
predicted = scaler.inverse_transform(predicted)

# Output
print("Predicted Temperature:")
print(predicted[:5])
```
